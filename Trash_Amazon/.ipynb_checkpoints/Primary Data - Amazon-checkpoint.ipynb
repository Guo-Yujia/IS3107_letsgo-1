```python
data = pd.read_csv("primary data.csv", 
                   dtype={'star_rating':int, 
                          'helpful_votes':int, 
                          'total_votes':int, 
                          'vine':int, 
                          'verified_purchase':int})
data['review_date'] = pd.to_datetime(data['review_date'])
data.sort_values(by=['review_date'], inplace=True)
data
```
