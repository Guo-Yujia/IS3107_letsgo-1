```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import requests
from spotipy.oauth2 import SpotifyOAuth
```

```python
credentials = json.load(open('authorization.json'))
client_id = credentials['client_id']
client_secret = credentials['client_secret']

playlist_index = 0

playlists = json.load(open('playlist_like_dislike.json'))
playlist_uri = playlists[playlist_index]['uri']
like = playlists[playlist_index]['like']
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=client_id,client_secret=client_secret)

sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
cats = {}
for i in range(0, 100, 50):
    cats.update(sp.categories(limit=50, offset=i))
```

```python
categories = sp.categories( limit=50, offset=0)
```

```python
pl = sp.category_playlists(category_id="toplists", limit=50, offset=0)
print(json.dumps(pl, indent=4))
```

```python
play=[]
for i, t in enumerate(categories["categories"]["items"]):
    try:
        pl = sp.category_playlists(category_id=t['id'], limit=50, offset=0)
        print('total of play list {}'.format(pl["playlists"]["total"]))
        total = int(pl["playlists"]["total"])
        if total <= 50:
            play.append(pl["playlists"]['items'][0])
        else:
            for j in range(0,total,50):
                play.append(sp.category_playlists(category_id=t['id'], limit=50, offset=j)["playlists"]['items'][0])
        print('len of play list {}'.format(len(play)))
    except:
        print(t['id'])
```
